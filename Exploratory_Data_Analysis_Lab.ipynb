# **Data Cleanup Using EDA Diagnostic, Statistical and Visualization Tools**

The following code:
- exemplifies the iterative approach to EDA and step-wise data cleanup processes
- should only be run once
- would benefit from a modular approach with more efficient and elegant function definition (e.g., lamda) to avoid runtime errors during the same sesion
- additional doucmentation for clarity when packaging and sharing with others

```python
# Convert the 'Age' column to integers, giving to float, Data last modified to date, etc.
df['ID'] = df['ID'].astype('int')
df['Age'] = df['Age'].astype(int)
df['Event 1 Attendance'] = df['Event 1 Attendance'].astype('bool')
df['Event 2 Attendance'] = df['Event 2 Attendance'].astype('bool')
df['Event 3 Attendance'] = df['Event 3 Attendance'].astype('bool')
df['Volunteer Activity 1'] = df['Volunteer Activity 1'].astype('bool')
df['Volunteer Activity 2'] = df['Volunteer Activity 2'].astype('bool')
df['Volunteer Activity 3'] = df['Volunteer Activity 3'].astype('bool')
df['Data Last Modified'] = pd.to_datetime(df['Data Last Modified'])
```

```python
df['Gift 1'].eq('').sum()
df['Gift 1'] = df['Gift 1'].replace('', np.nan)
df['Gift 1'] = df['Gift 1'].str.replace('[$,]', '', regex=True).astype(float)
df['Gift 1'] = df['Gift 1'].astype(float)

df['Gift 2'].eq('').sum()
df['Gift 2'] = df['Gift 2'].replace('', np.nan)
df['Gift 2'] = df['Gift 2'].str.replace('[$,]', '', regex=True).astype(float)
df['Gift 2'] = df['Gift 2'].astype(float)

df['Gift 3'].eq('').sum()
df['Gift 3'] = df['Gift 3'].replace('', np.nan)
df['Gift 3'] = df['Gift 3'].str.replace('[$,]', '', regex=True).astype(float)
df['Gift 3'] = df['Gift 3'].astype(float)

df['Total Giving History'].eq('').sum()
df['Total Giving History'] = df['Total Giving History'].replace('', np.nan)
df['Total Giving History'] = df['Total Giving History'].str.replace('[$,]', '', regex=True).astype(float)
df['Total Giving History'] = df['Total Giving History'].astype(float)
```

## Creating a data integrity score

```python
# reload example data via gsheets
url = 'https://docs.google.com/spreadsheets/d/1CJqX54aQSpP3MEtH8ADLI0nyKWHU2kHWtLcpc3qapic/export?format=csv'
df = pd.read_csv(url)
```

```python
# reapply cleanup demo scenario

# Convert the 'Age' column to integers, giving to float, Data last modified to date, etc.
df['ID'] = df['ID'].astype('int')
df['Age'] = df['Age'].astype(int)
df['Event 1 Attendance'] = df['Event 1 Attendance'].astype('bool')
df['Event 2 Attendance'] = df['Event 2 Attendance'].astype('bool')
df['Event 3 Attendance'] = df['Event 3 Attendance'].astype('bool')
df['Volunteer Activity 1'] = df['Volunteer Activity 1'].astype('bool')
df['Volunteer Activity 2'] = df['Volunteer Activity 2'].astype('bool')
df['Volunteer Activity 3'] = df['Volunteer Activity 3'].astype('bool')
df['Data Last Modified'] = pd.to_datetime(df['Data Last Modified'])

df['Gift 1'].eq('').sum()
df['Gift 1'] = df['Gift 1'].replace('', np.nan)
df['Gift 1'] = df['Gift 1'].str.replace('[$,]', '', regex=True).astype(float)
df['Gift 1'] = df['Gift 1'].astype(float)

df['Gift 2'].eq('').sum()
df['Gift 2'] = df['Gift 2'].replace('', np.nan)
df['Gift 2'] = df['Gift 2'].str.replace('[$,]', '', regex=True).astype(float)
df['Gift 2'] = df['Gift 2'].astype(float)

df['Gift 3'].eq('').sum()
df['Gift 3'] = df['Gift 3'].replace('', np.nan)
df['Gift 3'] = df['Gift 3'].str.replace('[$,]', '', regex=True).astype(float)
df['Gift 3'] = df['Gift 3'].astype(float)

df['Total Giving History'].eq('').sum()
df['Total Giving History'] = df['Total Giving History'].replace('', np.nan)
df['Total Giving History'] = df['Total Giving History'].str.replace('[$,]', '', regex=True).astype(float)
df['Total Giving History'] = df['Total Giving History'].astype(float)
```

```python
#num_missing_values = df.isna().sum().sum()
num_missing_values = df.isnull().sum().sum()

print("number of missing values:", num_missing_values)

num_records = df.shape[0]
print("number of records:", num_records)

data_integrity_score = (1 - (num_missing_values / num_records)) * 100

print("data integrity score:", round(data_integrity_score))
```

```python
num_missing_address = (df['Address'].isna()).sum()
print("number of missing addresses:", num_missing_address)

num_missing_comms = (df['Communication Preference'].isna()).sum()
print("number of missing communication preferences:", num_missing_comms)

num_missing_emails = (df['Email'].isna()).sum()
print("number of missing emails:", num_missing_emails)

num_records = df.shape[0]
print("number of record:",num_records)

data_integrity_score = (1 - ((num_missing_address + num_missing_comms + num_missing_emails) / num_records)) * 100

print("data integrity score = ", round(data_integrity_score))
```

## Refining your data integrity score

```python
# Refine data integrity score by dropping comms preferences, adding missing phone, duplicate rows, etc.

num_missing_address = (df['Address'].isna()).sum()
print("number of missing addresses:", num_missing_address)

num_missing_emails = (df['Email'].isna()).sum()
print("number of missing emails:", num_missing_emails)

num_missing_phones = (df['Phone'].isna()).sum()
print("number of missing phoness:", num_missing_phones)

num_duplicate_rows = df.duplicated().sum()
print("number of duplicate rows:", num_duplicate_rows)

num_records = df.shape[0]
print("number of records:", num_records)

data_integrity_score = (1 - ((num_missing_address + num_missing_emails + num_missing_phones + num_duplicate_rows) / num_records)) * 100


print("data integrity score:", round(data_integrity_score))
```
